## 1. What is an Array?

A NumPy array is a grid of values, all of the same type. This is different from Python lists, which can hold multiple types. This consistency allows NumPy to perform operations much faster.

* [dtype](https://numpy.org/doc/stable/reference/arrays.dtypes.html) - data type. Arrays always contain one type.
* [shape](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.shape.html) - dimensions of the data (e.g., 3×2).
* [data](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.data.html) - raw data storage in memory.

### Performance Comparison: Pure Python vs. NumPy

First, we define a task using standard Python lists.

````python
a = list(range(10000))
b = [ 0 ] * 10000
````

````python
%%timeit
for i in range(len(a)):
    b[i] = a[i]**2
````

Now, we perform the same task using [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html) and [vectorization](https://numpy.org/doc/stable/user/basics.types.html).

````python
import numpy as np
a = np.arange(10000)
b = np.zeros(10000)
````

````python
%%timeit
b = a ** 2
````

## 2. Creating Arrays

There are many ways to initialize arrays depending on the desired content.

````python
# Creating from lists
a = np.array([1,2,3])               
b = np.array([[1,2,3],[4,5,6]])     

print("Shape:", b.shape)
print("Size:", b.size)

# Built-in creators
print("Zeros:\n", np.zeros((2, 3)))
print("Ones:\n", np.ones((1,2)))
print("Full:\n", np.full((2,2),7))
print("Identity:\n", np.eye(2))
print("Linspace:\n", np.linspace(0,9,10))
````

### Array I/O and Datatypes

You can save arrays using [numpy.save](https://numpy.org/doc/stable/reference/generated/numpy.save.html) and convert types using [astype](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.astype.html).

````python
d = np.ones((3, 2), 'bool')
print("Original dtype:", d.dtype)
print("Converted to int:\n", d.astype('int'))

np.save('x.npy', a)
x_loaded = np.load('x.npy')
````

## 3. Array Maths and Vectorization

By default, basic arithmetic in NumPy is element-by-element. Unlike Matlab, NumPy uses `*` for element-wise multiplication and `@` for matrix multiplication.

* [numpy.add](https://numpy.org/doc/stable/reference/generated/numpy.add.html)
* [numpy.dot](https://numpy.org/doc/stable/reference/generated/numpy.dot.html)
* [numpy.transpose](https://numpy.org/doc/stable/reference/generated/numpy.transpose.html)

````python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print("Element-wise Add:\n", a + b)
print("Matrix Multiply (@):\n", a @ b)
print("Square Root:\n", np.sqrt(a))
````

## 4. Indexing and Slicing

Efficient extraction of data is key to NumPy's speed. You can select single elements, rows/columns, or apply conditions (Boolean indexing).

* [Indexing Documentation](https://numpy.org/doc/stable/user/basics.indexing.html)

````python
a = np.arange(16).reshape(4, 4)
print("Row 0:", a[0])
print("Col 0:", a[:,0])
print("Middle 2x2:\n", a[1:3, 1:3])

# Boolean Indexing
print("Values > 5:", a[a > 5])
````

## 5. Universal Functions (ufuncs) and Broadcasting

**ufuncs** are functions that operate element-by-element on arrays. They support **broadcasting**, which allows operations between arrays of different shapes.

* [ufunc Reference](https://numpy.org/doc/stable/reference/ufuncs.html)
* [Broadcasting Guide](https://numpy.org/doc/stable/user/basics.broadcasting.html)

````python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 10, 10])
print("Broadcasted addition:\n", a + b)

x = np.arange(12).reshape(3, 4)
print("Max along columns:", x.max(axis=0))
````
