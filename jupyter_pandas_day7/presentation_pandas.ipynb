## 1. A Quick Pandas Preview

We will use the Titanic passenger dataset to demonstrate the power of Pandas. We start by importing the library and loading the data directly from a URL into a [DataFrame](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html).

```python
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
titanic = pd.read_csv(url, index_col='Name')

# View the first 5 lines
titanic.head()
```

```python
# Print summary statistics
titanic.describe()
```

### Grouping and Visualizing
Pandas makes it easy to split data and apply functions. Here we look at survival by age group using [groupby](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html) and [hist](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html).

```python
print(titanic.groupby("Survived")["Age"].mean())

titanic.hist(column='Age', by='Survived', bins=25, figsize=(8,10),
             layout=(2,1), zorder=2, sharex=True, rwidth=0.9);
```

## 2. What's in a Dataframe?

A `DataFrame` is a tabular structure composed of rows and columns. Each column is a [Series](https://pandas.pydata.org/docs/reference/api/pandas.Series.html). Unlike NumPy, DataFrames can store mixed data types (integers, strings, floats) across different columns.

```python
# Inspect column info
titanic.info()

# Accessing a single column (Series)
titanic["Age"]

# Accessing index and columns
print(titanic.columns)
print(titanic.index)
```

### Selection and Indexing
You can select data by label ([loc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html)) or by integer position ([iloc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html)).

```python
titanic.loc['Lam, Mr. Ali', "Age"]          # Label-based
titanic.iloc[0:2, 3:6]                      # Integer-based

titanic.at['Lam, Mr. Ali', "Age"] = 42      # Fast scalar setter
titanic["is_passenger"] = True             # Set whole column

# Boolean indexing
titanic[titanic["Age"] > 70]
```

### Missing Data
Pandas uses `NaN` (Not a Number) for missing values. Key methods include [isna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isna.html), [dropna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html), and [fillna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html).

```python
titanic.isna()                    # Boolean mask
titanic.dropna(subset=["Cabin"])  # Drop rows missing Cabin info
titanic.fillna(0)                 # Fill missing with 0
```

## 3. Tidy Data

In tidy data, each column is a variable and each row is a measurement. Use [melt](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to transform "wide" data into "long" (tidy) data.

```python
runners = pd.DataFrame([
    {'Runner': 'Runner 1', 400: 64, 800: 128, 1200: 192, 1500: 240},
    {'Runner': 'Runner 2', 400: 80, 800: 160, 1200: 240, 1500: 300},
    {'Runner': 'Runner 3', 400: 96, 800: 192, 1200: 288, 1500: 360},
])

runners_tidy = pd.melt(runners, id_vars="Runner",
                       value_vars=[400, 800, 1200, 1500],
                       var_name="distance",
                       value_name="time")
```

## 4. Working with Dataframes

Data can be imported from many formats using [read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html), [read_excel](https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html), etc. You can also combine DataFrames using [concat](https://pandas.pydata.org/docs/reference/api/pandas.concat.html) or [merge](https://pandas.pydata.org/docs/reference/api/pandas.merge.html).

```python
import numpy as np
age = pd.DataFrame([
    {"Runner": "Runner 4", "Age": 18},
    {"Runner": "Runner 2", "Age": 21},
    {"Runner": "Runner 1", "Age": 23},
    {"Runner": "Runner 3", "Age": 19},
])

# Merge age into the runners table
runners_tidy.merge(age, on="Runner")
```

## 5. Time Series Superpowers

Pandas excels at handling date/time data. Use [to_datetime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) to convert columns and the `.dt` accessor for time properties.

```python
url_nobel = "https://github.com/AaltoSciComp/python-for-scicomp/raw/master/resources/data/laureate.csv"
nobel = pd.read_csv(url_nobel)

nobel["born"] = pd.to_datetime(nobel["born"], errors='coerce')
nobel["died"] = pd.to_datetime(nobel["died"], errors='coerce')
nobel["year"] = pd.to_datetime(nobel["year"], format="%Y")

print(nobel["born"].dt.day_name())

# Plotting by category
nobel["lifespan"] = (nobel["died"] - nobel["born"]).dt.days / 365
nobel.boxplot(column="lifespan", by="category")
```
